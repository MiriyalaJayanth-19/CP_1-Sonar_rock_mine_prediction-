```python
import pandas as pd
import numpy as np
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
```

```python
sonardf=pd.read_csv('sonar_data.csv',header=None)
```

```python
x=sonardf.drop(columns=60,axis=1)
y=sonardf[60]
```

```python
print(x)
```

```python
print(y)
```

```python
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=0.1,stratify=y,random_state=1)
```

```python
print(x.shape,xtrain.shape,ytrain.shape)
```

```python
ml=LogisticRegression()
```

```python
ml.fit(xtrain,ytrain)
```

```python
xtrainpred=ml.predict(xtrain)
train_accuracy=accuracy_score(xtrainpred,ytrain)
print(train_accuracy)
```

```python
xtestpred=ml.predict(xtest)
test_accuracy=accuracy_score(xtestpred,ytest)
print(test_accuracy)
```
